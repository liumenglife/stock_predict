```python
df_new = pd.DataFrame()
for idx, name in enumerate(df.name.unique()):
    print(idx, name)
    try:
        df_temp = df[df['name'] == name].sort_values(by=['date']).reset_index(drop=True)
        df_temp = Features.fnMACD(df_temp)
        df_temp = Features.fnBolingerBand(df_temp)
        df_temp = Features.fnRSI(df_temp)
        df_temp = Features.fnStoch(df_temp)
        df_temp = Features.change_prior_to(df_temp)
        df_temp = Features.fnMA(df_temp, m_N=[5, 20, 60, 120, 240])
    except Exception as e:
        print(e)
        continue
    df_new = pd.concat([df_new, df_temp])

df = df_new.reset_index(drop=True)


# df_new = pd.DataFrame()
# for idx, name in enumerate(df.name.unique()):
#     df_new = pd.concat([df_new, Features.fnMACD(df[df['name'] == name].sort_values(by=['date']).reset_index(drop=True))])

# df = df_new.reset_index(drop=True)



# df_new = pd.DataFrame()
# for idx, name in enumerate(df.name.unique()):
#     df_new = pd.concat([df_new, Features.fnBolingerBand(df[df['name'] == name].sort_values(by=['date']).reset_index(drop=True))])

# df = df_new.reset_index(drop=True)



# df_new = pd.DataFrame()
# for idx, name in enumerate(df.name.unique()):
#     df_new = pd.concat([df_new, Features.fnRSI(df[df['name'] == name].sort_values(by=['date']).reset_index(drop=True))])

# df = df_new.reset_index(drop=True)


# df_new = pd.DataFrame()
# for idx, name in enumerate(df.name.unique()):
#     df_new = pd.concat([df_new, Features.fnStoch(df[df['name'] == name].sort_values(by=['date']).reset_index(drop=True))])

# df = df_new.reset_index(drop=True)




# df_new = pd.DataFrame()
# for idx, name in enumerate(df.name.unique()):
#     df_new = pd.concat([df_new, Features.change_prior_to(df[df['name'] == name].sort_values(by=['date']).reset_index(drop=True))])

# df = df_new.reset_index(drop=True)



# df_new = pd.DataFrame()
# for idx, name in enumerate(df.name.unique()):
#     df_new = pd.concat([df_new, Features.fnMA(df[df['name'] == name].sort_values(by=['date']).reset_index(drop=True), m_N=[5, 20, 60, 120, 240])])

# df = df_new.reset_index(drop=True)
```
